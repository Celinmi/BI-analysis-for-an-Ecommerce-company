```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('users_data.csv') 
```

```python
def change_phone(country,phone):
    dic_c_c={
    "France": "01833",
    "Spain": "01834",
    "Israel": "00972",
    "Germany": "01849",
    "United Kingdom": "01844",
    "Italy": "01839",
    "United States": "0181",
    "Australia": "01861",
    }
    if type(phone)== str:
        for key,val in dic_c_c.items():
            if country == key:
                if phone[0] != '+':
                    phone= val+phone
        # Remove characters that are neither numbers nor "+"
        phone = re.sub(r'[^0-9+]', '', phone)
        phone = phone.replace(" ", "")
    return phone
```

```python
lst_good_phone=[]
for i in range(df.shape[0]):
    lst_good_phone.append(change_phone(df.loc[i,'Country'],df.loc[i,'PhoneNumber']))
df['PhoneNumber'] = lst_good_phone
df
```
